# Data Visualization

```python
# Sample data
data = {
    'VehicleClass': ['Midsize Cars', 'Compact Cars',  'Subcompact Cars',  'Large Cars', 'Minicompact Cars'],
    'CombinedMPG': [18.74, 25.22, 28.68, 35.43, 17.67]
}
fuel_econ_Clean = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
fuel_econ_Clean.plot(x= 'VehicleClass', y= 'CombinedMPG', kind='bar', legend=False)
plt.title('Average Combined Fuel Efficiency by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('Average Combined MPG')
# plt.xticks(rotation = 45)

plt.show()
```

```python
# Sample data
data = {
    'Year': [2013, 2014, 2015, 2016],
    'CombinedMPG': [18.74, 20.22, 22.67, 24.68]
}
fuel_econ_Clean = pd.DataFrame(data)

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(fuel_econ_Clean['Year'], fuel_econ_Clean['CombinedMPG'], marker='o', linestyle='-', color='green')
plt.title('Yearly Average Combined Fuel Economy')
plt.xlabel('Year')
plt.ylabel('Average Combined MPG')
plt.grid(True)
plt.show()
```

```python
# Sample data
data = {
    'CombinedMPG': [18.74, 25.22, 20.67, 28.68, 22.56]
}
fuel_econ_Clean = pd.DataFrame(data)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(fuel_econ_Clean['CombinedMPG'], bins=5, edgecolor='black', alpha=0.7)
plt.title('Distribution of Combined Fuel Economy')
plt.xlabel('Combined MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sample data
data = {
    'Transmission': ['Automatic (AM6)', 'Automatic (AM-S6)', 'Automatic (S6)', 'Automatic 7-spd'],
    'CombinedMPG': [18.74, 25.22, 20.67, 28.68]
}
fuel_econ_Clean = pd.DataFrame(data)

# Create the box plot
plt.figure(figsize=(10, 6))
sb.boxplot(x='Transmission', y='CombinedMPG', data=fuel_econ_Clean)
plt.title('Fuel Economy Comparison by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Combined MPG')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sample data
data = {
    'Make': ['Nissan', 'Volkswagen', 'Hyundai', 'Chevrolet'],
    'CO2_emissions': [471, 349, 429, 310]
}
fuel_econ_Clean = pd.DataFrame(data)

# Calculate average CO2 emissions by make
avg_co2_by_make = fuel_econ_Clean.groupby('Make')['CO2_emissions'].mean().reset_index()

# Create the bar chart
plt.figure(figsize=(10, 6))
avg_co2_by_make.plot(x='Make', y='CO2_emissions', kind='bar', legend=False)
plt.title('Average CO2 Emissions by Vehicle Make')
plt.xlabel('Vehicle Make')
plt.ylabel('Average CO2 Emissions (grams/mile)')
plt.xticks(rotation=45)
plt.show()
```
